```python
#for data frames and easy data manipulation
import pandas as pd
import numpy as np
#to remove all the most common words from the body of the wikipedia pages, and the titles
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
#to count occurances of words
from collections import Counter
#for string functions
import string
#for word stemming
from nltk.stem import PorterStemmer, WordNetLemmatizer
porter_stemmer = PorterStemmer()
```

```python
print(stop)
```

```python
#import data file that came from web scraper
df = pd.read_json('/content/drive/MyDrive/Colab Notebooks/cswiki1000.json')
#Time to do some data processing. Get rid of all new lines, get rid of all punctuation, make everything lowercase. 
#get rid of all the new lines.
df = df.replace(r'\n',' ', regex=True)
#removing any extra spaces between words
df = df.apply(lambda x: x.str.strip())
#make everything lowercase
df['body'] = df['body'].str.lower()
#remove punctuation 
df['body']= df['body'].str.replace('[^\w\s]','')
```

```python
#remove stop words from title and body
df['body'] = df['body'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
#create function to use our stemmer.
def stem_sentences(sentence):
    tokens = sentence.split()
    stemmed_tokens = [porter_stemmer.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)
#vall function on body
df['body'] = df['body'].apply(stem_sentences)
```

```python
#display most 50 most common words in body
Counter(" ".join(df["body"]).split()).most_common(50)
```
